### Faixa Etária de Demitidos

```python
bins = [18, 25, 38, 60]
labels = ['18 - 25', '26 - 38', '39 - 60']
df['Faixa Etária'] = pd.cut(df_demitidos['Idade'], bins=bins, labels=labels, right=True)

# Contagem de usuários por categoria
atividade_counts = df['Faixa Etária'].value_counts().reindex(labels)
atividade_percent = (atividade_counts / atividade_counts.sum()) * 100

# Criando o gráfico de barras
plt.figure(figsize=(8,5))
bars = plt.bar(atividade_counts.index, atividade_counts.values, color=['red', '#E8710A', '#FEC724', 'green'])
plt.title('Distribuição por Faixa Etária dos funcionários que pediram demissão')
plt.xticks(rotation=30)

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)

plt.gca().get_yaxis().set_visible(False)

plt.ylim(0, max(atividade_counts.values) + 20)

for bar, pct in zip(bars, atividade_percent):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height/2, f'{pct:.1f}%', ha='center', va='center', fontsize=12, color='black', fontweight='bold')

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar os dados
url = 'https://docs.google.com/spreadsheets/d/e/2PACX-1vSbKntUfJi-sTMZV0fRT8QsVwLx8wBVBjKvsa5OblXc0qSIKSL0xaEgNnxJC9VEVUl9K1SRQmBoSgKV/pub?gid=602564934&single=true&output=csv'
df = pd.read_csv(url)

# Remover espaços extras dos nomes das colunas
df.columns = df.columns.str.strip()

# Verificar se a coluna "Demissão" existe no dataset
if 'Demissão' in df.columns:
    # Manter apenas os funcionários demitidos
    df_demitidos = df[df['Demissão'] == 'Sim'].copy()

    # Selecionar apenas colunas numéricas
    df_numerico = df_demitidos.select_dtypes(include=['number'])

    # Calcular a matriz de correlação
    correlation_matrix = df_numerico.corr()

    # Selecionar os 10 fatores mais relevantes (com maior correlação entre si)
    top_10 = correlation_matrix.abs().mean().nlargest(10).index

    # Criar heatmap com os 10 dados mais relevantes
    plt.figure(figsize=(12, 8))
    sns.heatmap(df_numerico[top_10].corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
    plt.title("Matriz de Correlação - Apenas Funcionários Demitidos")
    plt.show()

else:
    print("⚠️ A coluna 'Demissão' não foi encontrada no dataset.")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar os dados
url = 'https://docs.google.com/spreadsheets/d/e/2PACX-1vSbKntUfJi-sTMZV0fRT8QsVwLx8wBVBjKvsa5OblXc0qSIKSL0xaEgNnxJC9VEVUl9K1SRQmBoSgKV/pub?gid=602564934&single=true&output=csv'
df = pd.read_csv(url)

# Remover espaços extras dos nomes das colunas
df.columns = df.columns.str.strip()

# Verificar se a coluna "Demissão" existe no dataset
if 'Demissão' in df.columns:
    # Manter apenas os funcionários demitidos
    df_demitidos = df[df['Demissão'] == 'Sim'].copy()

    # Selecionar apenas colunas numéricas
    df_numerico = df_demitidos.select_dtypes(include=['number'])

    # Calcular a matriz de correlação
    correlation_matrix = df_numerico.corr()

    # Selecionar os 10 fatores mais relevantes (com maior correlação entre si)
    top_10 = correlation_matrix.abs().mean().nlargest(10).index

    # Criar heatmap com os 10 dados mais relevantes
    plt.figure(figsize=(12, 8))
    sns.heatmap(df_numerico[top_10].corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
    plt.title("Matriz de Correlação - Apenas Funcionários Demitidos")
    plt.show()

else:
    print("⚠️ A coluna 'Demissão' não foi encontrada no dataset.")
```

```python
plt.figure(figsize=(8, 6))

sns.scatterplot(data=df_demitidos, x='Anos trabalhando com o atual gestor', y='Anos desde a última promoção')

plt.title('Anos trabalhando com o atual gestor vs. Anos desde a última promoção (Funcionários que pediram demissão)')
plt.xlabel('Anos trabalhando com o atual gestor')
plt.ylabel('Anos desde a última promoção')
plt.show()
```

```python
plt.figure(figsize=(12, 8))

df_contratados = df[df['Demissão'] == 'Não'].copy()
df_demitidos = df[df['Demissão'] == 'Sim'].copy()

sns.scatterplot(data=df_contratados, x='Total de anos trabalhados na empresa', y='Anos desde a última promoção', label='Contratados', alpha=0.6)
sns.scatterplot(data=df_demitidos, x='Total de anos trabalhados na empresa', y='Anos desde a última promoção', label='Pediram demissão', alpha=0.6)

plt.title('Tempo na Empresa vs. Tempo Sem Promoção (Contratados vs. Pediram demissão)')
plt.xlabel('Total de anos trabalhados na empresa')
plt.ylabel('Anos desde a última promoção')
plt.legend()
plt.show()
```
